# Data Visualization-Create a histogram for each feature in the dataset to illustrate the feature distributions. Plot each histogram.


```python
dt['Embarked'].value_counts()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt



fig,axes = plt.subplots(3,2,figsize=(18,12))

sns.histplot(data=dt,x='Survived',ax=axes[0][0],bins=5,kde=True,hue='Embarked',alpha=0.5)
sns.histplot(data=dt,x='Pclass',ax=axes[0][1],bins=5,kde=True,hue='Embarked',alpha=0.5)
sns.histplot(data=dt,x='SibSp',ax=axes[1][0],bins=5,kde=True,hue='Embarked',alpha=0.5)
sns.histplot(data=dt,x='Parch',ax=axes[1][1],bins=5,kde=True,hue='Embarked',alpha=0.5)
sns.histplot(data=dt,x='Age',ax=axes[2][0],bins=5,kde=True,hue='Embarked',alpha=0.5)
sns.histplot(data=dt,x='Fare',ax=axes[2][1],bins=5,kde=True,hue='Embarked',alpha=0.5)
```

```python
fig,axes = plt.subplots(3,2,figsize=(18,12))
sns.boxplot(data=dt,y='Survived',x='Embarked',ax=axes[0][0])
sns.boxplot(data=dt,y='Pclass',x='Embarked',ax=axes[0][1])
sns.boxplot(data=dt,y='SibSp',x='Embarked',ax=axes[1][0])
sns.boxplot(data=dt,y='Parch',x='Embarked',ax=axes[1][1])
sns.boxplot(data=dt,y='Age',x='Embarked',ax=axes[2][0])
sns.boxplot(data=dt,y='Fare',x='Embarked',ax=axes[2][1])
```
